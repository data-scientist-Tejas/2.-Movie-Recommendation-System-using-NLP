```python
import pandas as pd
```

```python
df=pd.read_csv("tmdb_5000_credits.csv")
```

```python
df
```

```python
import ast #allows you to parse, analyze, and manipulate Python code programmatically
```

```python
def convert3(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i["name"])
    return l
```

```python
df['cast']=df['cast'].apply(convert3)
```

```python
df
```

```python
def fetch_director(text):
    l=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            l.append(i['name'])
    return l
```

```python
df['crew']=df['crew'].apply(fetch_director)
```

```python
df
```

```python
df['tag']=df['cast']+df['crew']
```

```python
df=df.drop(columns=['cast','crew'])
```

```python
df
```

```python

df['tag'] = df['tag'].apply(lambda x: " ".join(x) if isinstance(x, list) else str(x))
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(df['tag']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
df[df['title']=='The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = df[df['title'] == movie].index[0]  # Find the index of the movie
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])  # Sort movies by similarity
    for i in distances[1:6]:  # Loop through the top 5 most similar movies
        print(df.iloc[i[0]].title)  # Print the titles of those movies
```

```python
#enumerate(similarity[index]): This adds an index to each similarity score, so we can know which movie each similarity score corresponds
```

```python
recommend('Gandhi')
```

```python
recommend('Gandhi')
```

```python
recommend('Spectre')
```

```python
recommend('Shanghai Calling')
```
